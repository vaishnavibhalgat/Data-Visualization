# 4. Generate a dataset with categories and numerical values. Visualize the distribution of a numerical variable across different categories.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate Dataset
np.random.seed(42)
categories = ['Category A', 'Category B', 'Category C', 'Category D']
data = {
    'Category': np.random.choice(categories, 100),  # Randomly assign categories
    'Value': np.random.normal(50, 10, 100)         # Random numerical values
}
df = pd.DataFrame(data)

# Step 2: Visualize Distribution with Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')
plt.title("Distribution of Values Across Categories (Box Plot)", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Step 3: Visualize Distribution with Violin Plot
plt.figure(figsize=(8, 5))
sns.violinplot(x='Category', y='Value', data=df, palette='Set3')
plt.title("Distribution of Values Across Categories (Violin Plot)", fontsize=14)
plt.xlabel("Categories", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate Synthetic Dataset with Correlated Features
np.random.seed(42)

# Generate correlated features
x = np.random.rand(100)
y = x + np.random.normal(0, 0.1, 100)  # y is highly correlated with x
z = np.random.rand(100)  # Independent feature

# Create a DataFrame
df = pd.DataFrame({'Feature X': x, 'Feature Y': y, 'Feature Z': z})

# Step 2: Calculate the Correlation Matrix
corr_matrix = df.corr()

# Step 3: Visualize the Correlation Matrix using Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=1, vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap", fontsize=14)
plt.show()
```

```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go

# Step 1: Generate the x, y, and z data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Step 2: Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Step 3: Create the 3D Surface Plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])

# Step 4: Customize the plot and display
fig.update_layout(
    title='3D Surface Plot of the Function z = sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    autosize=True
)

# Show the plot
fig.show()
```

# 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Step 1: Generate random data
np.random.seed(10)

n = 100  # Number of points
x = np.random.random(n) * 10  # Random x values between 0 and 10
y = np.random.random(n) * 10  # Random y values between 0 and 10
sizes = np.random.randint(5, 50, size=n)  # Random sizes for markers
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], size=n)  # Random colors

# Step 2: Create the scatter plot
output_notebook()  # Display the plot inline in the Jupyter notebook

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(
    title="Random Scatter Plot with Varying Marker Sizes and Colors",
    x_axis_label='x',
    y_axis_label='y',
    width=800,
    height=400  # Corrected attributes
)

p.scatter('x', 'y', size='sizes', color='colors', legend_field='colors', source=source, fill_alpha=0.6)

# Step 3: Show the plot
show(p)
```

# 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Display plots inline in the notebook
output_notebook()

# Generate data
data_hist = np.random.randn(1000)

# Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure
p = figure(
    title="Histogram of Data Distribution",
    x_axis_label="Value",
    y_axis_label="Frequency",
    width=800,
    height=400
)

# Add a quad glyph to represent the histogram
p.quad(
    top=hist,
    bottom=0,
    left=edges[:-1],
    right=edges[1:],
    fill_color="skyblue",
    line_color="white",
    alpha=0.7
)

# Customize the chart
p.xgrid.grid_line_color = None
p.y_range.start = 0

# Show the plot
show(p)
```

# 5. Create a Bokeh heatmap using the provided dataset.

 y = np.linspace(0, 1, 10)
 xx,

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource  # Import ColumnDataSource

# Display plots inline in the notebook
output_notebook()

# Generate data for heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Flatten the data for plotting
x_flat = xx.flatten()
y_flat = yy.flatten()
values = data_heatmap.flatten()

# Create a color mapper for the heatmap
color_mapper = LinearColorMapper(palette="Viridis256", low=np.min(values), high=np.max(values))

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, values=values)) # Create a ColumnDataSource

# Create a figure
p = figure(
    title="Heatmap of Random Data",
    x_axis_label="X-axis",
    y_axis_label="Y-axis",
    width=800,
    height=400,
    tools="pan,box_zoom,reset"
)

# Add rectangles for the heatmap, referencing columns from the source
p.rect(
    x='x',  # Reference the 'x' column from the source
    y='y',  # Reference the 'y' column from the source
    width=0.1,
    height=0.1,
    fill_color=linear_cmap("values", palette="Viridis256", low=np.min(values), high=np.max(values)),
    line_color=None,
    source=source  # Pass the ColumnDataSource
)

# Add a color bar to the plot
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0), title="Intensity")
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
